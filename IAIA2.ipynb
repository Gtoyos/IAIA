# Pretrained models

```python
#required imports
import torch, torchvision
from tqdm.notebook import tqdm as tqdm
import matplotlib.pyplot as plt
import gzip
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import os
import zlib

from torchvision.models import resnet50, mobilenet_v2, ResNet50_Weights
```

```python
#get weights
def get_weights(model):
    weights = []
    for param in model.parameters():
        if param.data.dim() == 4:
            w = param.data.detach().cpu().numpy()
            #round to 4 decimal places
            w = np.round(w, 4)
            weights.append(w)
        #weights.append(param.data.detach().cpu().numpy())
    return weights
```

```python
trained = []
no_trained = []
model = mobilenet_v2(weights="IMAGENET1K_V1")
weights = get_weights(model)
print(weights[0])
for w in weights:
    cp = len(zlib.compress(w, level=9))
    ncp = len((zlib.compress(w, level=0)))
    trained.append(cp/ncp)
model = mobilenet_v2(weights=None)
weights = get_weights(model)
#print(weights[0])
for w in weights:
    cp = len(zlib.compress(w, level=9))
    ncp = len((zlib.compress(w, level=0)))
    no_trained.append(cp/ncp)

for i in range(len(trained)):
    print(no_trained[i]-trained[i])
```

```python
model = torchvision.models.resnet18(pretrained=False)
weights = get_weights(model)
for w in weights:
    cp = len(zlib.compress(w, level=9))
    ncp = len((zlib.compress(w, level=0)))
    print(cp/ncp)
```

```python
#get pretrained model
model = torchvision.models.resnet18(pretrained=True)

net_ratio = []
for i in range(len(get_weights(model))):
    net_ratio.append([])


for i,w in enumerate(get_weights(model)):
    f = gzip.GzipFile('compress.gz', "w")
    np.save(file=f, arr=w)
    f.close()
    np.save('normal.npy', w)
    compressed_size = os.path.getsize('compress.gz')
    normal_size = os.path.getsize('normal.npy')
    net_ratio[i].append(compressed_size/normal_size)
    
    os.remove('compress.gz')
    os.remove('normal.npy')
```
